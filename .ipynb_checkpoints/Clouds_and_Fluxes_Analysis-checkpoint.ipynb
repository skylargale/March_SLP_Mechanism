### Cloud Cover

```python
ds = xr.open_dataset('/pscratch/sd/s/skygale/snow/data/CERES_EBAF-TOA_Ed4.2.1_Subset_200003-202412.nc')
ds
```

```python
# Arctic mask: lat ≥ 70°N
arctic_ds = ds.sel(lat=ds.lat >= 70)

# Convert time to numeric for regression
time_num = (arctic_ds.time.dt.year + (arctic_ds.time.dt.month - 1)/12).values

# Arctic spatial average
weights = np.cos(np.deg2rad(arctic_ds.lat))  # area weights
weights.name = "weights"

# Arctic-mean time series
toa_net_avg = arctic_ds['toa_net_all_mon'].weighted(weights).mean(dim=('lat', 'lon'))
time_num = (arctic_ds.time.dt.year + (arctic_ds.time.dt.month - 1)/12).values

def bootstrap_slope(x, y, n=1000):
    rng = np.random.default_rng(42)
    slopes = []
    for _ in range(n):
        idx = rng.choice(len(x), size=len(x), replace=True)
        x_sample = x[idx]
        y_sample = y[idx]
        if np.all(np.isfinite(y_sample)):
            slope, _, _, _, _ = linregress(x_sample, y_sample)
            slopes.append(slope)
    return np.array(slopes)

slopes = bootstrap_slope(time_num, toa_net_avg.values, n=1000)
slope_mean = slopes.mean()
slope_std = slopes.std()

# Compute Arctic trend time series
slope, intercept, r, p, stderr = linregress(time_num, toa_net_avg.values)
trend_line = slope_mean * time_num + (toa_net_avg.mean().item() - slope_mean * time_num.mean())

plt.figure(figsize=(8, 4))
plt.plot(arctic_ds.time, toa_net_avg, label='Arctic Mean')
plt.plot(arctic_ds.time, trend_line, 'r', label=f'Trend: {slope_mean:.2f} ± {2*slope_std:.2f} W/m²/yr')
plt.title('Arctic TOA Net Flux Trend (±2σ)')
plt.ylabel('W/m²')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def compute_spatial_trend(data, time_num):
    slope_map = xr.full_like(data.isel(time=0), fill_value=np.nan)
    
    for i in range(data.lat.size):
        for j in range(data.lon.size):
            y = data[:, i, j].values
            if np.isfinite(y).all():
                slope, *_ = linregress(time_num, y)
                slope_map[i, j] = slope
    return slope_map

toa_net_trend_map = compute_spatial_trend(arctic_ds['toa_net_all_mon'], time_num)

plt.figure(figsize=(10, 5))
ax = plt.subplot(1, 1, 1, projection=ccrs.NorthPolarStereo())
ax.set_extent([-180, 180, 50, 90], crs=ccrs.PlateCarree())
ax.coastlines()

cf = ax.contourf(
    arctic_ds.lon, arctic_ds.lat, toa_net_trend_map,
    transform=ccrs.PlateCarree(), cmap='coolwarm', levels=21, extend='both'
)

plt.title('TOA Net Flux Trend (W/m²/year)')
plt.colorbar(cf, orientation='horizontal', pad=0.05)
plt.tight_layout()
plt.show()
```
