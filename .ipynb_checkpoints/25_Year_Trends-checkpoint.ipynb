```python
import xarray as xr
import glob
import numpy as np
import xesmf as xe
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
```

### Load data

```python
# Load data
hist = glob.glob('/pscratch/sd/s/skygale/monthly-trends/hist/arctic/*')
spliced = glob.glob('/pscratch/sd/s/skygale/monthly-trends/spliced/arctic/*')
sat = glob.glob('/pscratch/sd/s/skygale/monthly-trends/obs/sat/*')

mod_data = []
for path in spliced:
    ds = xr.open_dataset(path).X_SAT_SLP.isel(period=-2)
    mod_data.append(np.nanmean(ds, axis=0))
mod_data = np.nanmean(mod_data, axis=0)
mod_sat, mod_slp = mod_data[:, :, :, 0], mod_data[:, :, :, 1]

sat_data = []
for path in sat:
    ds = xr.open_dataset(path).DATA
    sat_data.append(ds)
obs_sat = np.nanmean(sat_data, axis=0)

np.shape(mod_sat), np.shape(mod_slp), np.shape(obs_sat)
```

```python
# Load full dataset and decode time
ds = xr.open_dataset('/pscratch/sd/s/skygale/monthly-trends/obs/slp/21267e161018ad8fac3c32ebc48d22db.nc')
ds = xr.decode_cf(ds)

# Optionally rename 'valid_time' to 'time'
ds = ds.rename({'valid_time': 'time'})

# Select MSL and time slice
ds_sel = ds.msl.sel(time=slice('2000-01', '2024-12'))

# Remove monthly slp climatology over given period
slp_clim = ds_sel.groupby('time.month').mean('time')
slp_clean = ds_sel.groupby('time.month') - slp_clim

# Monthly trends
group_month = slp_clean.groupby('time.month')
month_trend_fit = []
for month in group_month.groups.keys():
    ds_month = np.array(group_month[month])
    month_to_fit = np.reshape(ds_month, (ds_month.shape[0], ds_month.shape[1]*ds_month.shape[2]))
    month_trend = np.polyfit(np.arange(0, 2.5, 1/10), month_to_fit, 1)[0]
    month_trend_fit.append(np.reshape(month_trend, (ds_month.shape[1:])))
month_trends = np.stack(month_trend_fit, 0)

# Create output grid as xarray Dataset
lats = np.arange(-88.75, 88.751, 2.5)
lons = np.arange(1.25, 358.751, 2.5)

output_grid = xr.Dataset({
    'lat': (['lat'], lats),
    'lon': (['lon'], lons),
})

# Regrid using xESMF
regridder = xe.Regridder(ds, output_grid, method="bilinear", periodic=True)

# Apply the regridder
obs_slp = regridder(month_trends)

print(np.shape(obs_slp))
```

### Plot 25 year trends

```python
month_names = [
    'January', 'February', 'March',
    'April', 'May', 'June', 'July',
    'August', 'September', 'October',
    'November', 'December'
]
```

```python
fig, axes = plt.subplots(
    3, 4, figsize=(12, 8), dpi=300,
    subplot_kw={'projection': ccrs.NorthPolarStereo()},
    constrained_layout=True
)

axes = axes.ravel()
for i, ax in enumerate(axes):    
    cf = ax.contourf(
        lons, lats[44:], mod_sat[i],
        transform=ccrs.PlateCarree(),
        cmap='RdBu_r',
        levels=np.linspace(-0.2, 0.2, 0.01),
        extend='both'
    )

    # Set map extent and features
    ax.set_extent([-180, 180, 60, 90], ccrs.PlateCarree())
    ax.coastlines()
    ax.set_title(f'{month_names[i]}', fontsize=10)

    # Gridlines
    gl = ax.gridlines(draw_labels=False, linewidth=0.3, color='gray', alpha=0.5, linestyle='--')
    gl.xlocator = plt.MultipleLocator(60)
    gl.ylocator = plt.FixedLocator([60, 80])

# Colorbar
cbar = fig.colorbar(cf, ax=axes.ravel(), location='right', shrink=0.85, aspect=20, pad=0.02)
cbar.set_ticks([-0.2, -0.15, -0.1, -0.05, 0, 0.05, 0.1, 0.15, 0.2])
cbar.set_ticklabels(['-0.2', '-0.15', '-0.1', '-0.05', '0', '0.05', '0.1', '0.15', '0.2'])
cbar.ax.minorticks_off()
cbar.set_label('(decade$^{-1}$)', fontsize=14)

plt.suptitle("MMM Monthly SAT Trends (2000–2024)", fontsize=16, y=1.03)
plt.show()
```
